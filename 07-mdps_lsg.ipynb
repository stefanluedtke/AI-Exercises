# MDPs
Gegeben sei die folgende Umgebung mit 4 Räumen, die durch Türen verbunden sind. Die
Räume sind von 0 bis 3 nummeriert und 4 bedeutet draußen. Ein Roboter soll, nachdem er
in einigem beliebigen Raum positioniert wurde, durch Q-learning den Weg nach draußen
finden, d.h., 4 (draußen) ist der Zielzustand. Alle Türen die nach draußen führen haben
ein Reward von 100 und alle anderen 0.

1. Modellieren Sie die Räume als einen Graph, in dem die Räume als Zustände (Knoten) und die Türverbindungen als Transitionen gegeben sind. Notieren Sie die Rewards für jede Transition.
2. Geben Sie die Reward-Matrix R, die die Rewards zwischen den Zustände beschreibt. Unmögliche Transitionen bekommen ein Reward von -1.
3. Geben Sie die initiale Q-Matrix an! Nehmen Sie dabei an, dass der Roboter die Anzahl der Räume kennt.
4. Berechnen Sie die Rewards für eine “Episode”, die im Raum 1 anfängt. Benutzen Sie den  discount factor $\gamma = 0.5$. Benutzen Sie dafür die folgende Formel (diese resultiert aus der Standardregel für die Lernrate $\alpha = 1$): $Q(state, action) = R(state, action) + \gamma ∗ max[Q(next state, all actions)]$. Dabei sollen die Aktionen so gewählt werden, das der Roboter die Räume in der folgenden Reihenfolge besucht: 1 → 3 → 1 → 2 → 4. Geben Sie die resultierende Q-Matrix an!

## Ontologien
Mario ist Pizza-Bäcker (und Informatiker) und möchte die Pizzen, die in seinem Restaurant
angeboten werden, in einer Ontologie darstellen. Er hat folgende Vorstellung von den
Bestandteilen der Pizzen: Es gibt verschiedene Arten von Pizzen, je nach Boden und
Belag. Es gibt normale Böden, dünne Böden und dicke Böden. Es gibt vegetarische Beläge,
Käse-Beläge, Saucen (zählen auch als Belag), und Fleisch-Beläge. Eine Pizza, die einen
dünnen Boden hat, wird als italienische Pizza bezeichnet. Eine vegetarische Pizza hat nur
vegetarische Beläge und Saucen-Beläge. Eine Pizza, die Käse-Belag hat, wird als Käse-
Pizza bezeichnet.
Diese Aufgabe soll mit Protégé bearbeitet werden, das Sie sich unter http://protege.stanford.edu/ herunterladen können.
1. Definieren Sie die Klassen Pizza, Boden, und Belag.
2. Definieren Sie die Oject-Properties hatBoden und hatBelag.
3. Definieren Sie die Klassen ItalienischePizza, KäsePizza und VegetarischePizza entsprechend.
4. Probieren Sie den Reasoner aus, der mit Protégé mitgeliefert wird. Versuchen Sie, eine Pizza-Klasse zu definieren, die sowohl eine Subklasse von VegetarischePizza als auch von ItalienischePizza ist. Was gibt der Reasoner für diese Klasse aus?
5. Visualisieren Sie Ihre Ontologie mit OntoGraf (Window → Tabs → OntoGraf).

Eine ausführliche Lösung finden Sie hier: http://owl.cs.manchester.ac.uk/publications/talks-and-tutorials/protg-owl-tutorial/
