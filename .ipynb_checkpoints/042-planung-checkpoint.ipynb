# Planung

## 1. Ein einfaches Planungs-Problem
Das Anziehen von Socken und Schuhen ist ein Planungsproblem. Schuhe können nur angezogen werden, wenn die jeweilige Socke schon angezogen ist. In dieser Aufgabe soll das Anziehen von Socken und Schuhen modelliert werden.
* Zeichnen Sie den partiell geordneten Plan, und alle total geordneten Pläne auf.
* Lösen Sie das Problem mit dem ff Planner. Laden Sie dafür die Datei “Uebung42.zip” herunter und entpacken Sie diese. Den Planer können Sie wie folgt aufrufen: ./ff -o shoes/d.pddl -f shoes/p.pddl. Interpretieren Sie die Ausgabe. Alternativ können Sie auch den Online-Planer auf dieser Website verwenden: http://editor.planning.domains

## 2. Das Office-Planungsproblem
In dieser Aufgabe soll der typische Arbeitstag der Mitarbeiter am Lehrstuhl MMIS (Kaffee holen, umherlaufen, Dinge ausdrucken) als Planungs-Problem modelliert werden. Mitarbeiter können nur Kaffee machen, wenn die Kaffeemaschine mit Kaffee und Wasser befüllt ist. Kaffee kann am Kaffee-Glas geholt werden, Wasser am Wasserhahn. Mitarbeiter können
immer nur ein Ding gleichzeitig in der Hand halten. Am Drucker kann nur gedruckt werden, wenn der Drucker mit Papier befüllt ist. Papier gibt es am Papier-Schrank. Das Ziel der Mitarbeiter ist es, alle Printjobs gedruckt zu haben, alle einen Kaffee in der Hand zu halten, und am Ausgang zu stehen.



Die verfügbare Aktionen der Mitarbeiter sind:

* goto: Ein Mitarbeiter bewegt sich zwischen zwei Positionen. Die beiden Positionen dürfen nicht gleich sein.
* fetch: Ein Mitarbeiter nimmt Papier, Kaffee oder Wasser, wenn er sich an der jeweigen Location befindet.
* replenish: Ein Mitarbeiter füllt den Drucker oder die Kaffeemaschine auf. Das geht dann, wenn der die entsprechende Ressource in der Hand hält.
* get-coffee: Ein Mitarbeiter nimmt sich einen Kaffee, wenn er an der Maschine steht, und diese befüllt ist.
* print: Ein Mitarbeiter druckt einen Printjob, wenn er am Drucker steht, und der Drucker Papier enthält.

Definieren Sie die entsprechenden Aktionen in der Datei abc-d.pddl. Rufen Sie den Planer wie folgt auf: ./ff -o abc/abc-d.pddl -f abc/abc-p1.pddl. Interpretieren Sie die Ausgabe
